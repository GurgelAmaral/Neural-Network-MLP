#Modelo orientado a objetos

```python
import numpy as np
import random
import warnings

#Classe que define a arquitetura da rede e seus hiperparâmetros
class NeuralNetwork:
    def __init__(self, layer_neurons, learning_rate, num_epochs):
        """
        :param layer_neurons: Lista onde cada elemento representa o número de neurônios em uma camada.
                  EX: [2, 2, 1] => 1a camada tem 2 neuronios; 2a camada tem 2 neuronios; ultima camada tem 1 neuronio
        :param learning_rate: Taxa de aprendizado de (0, 1].
        :param num_epochs: Número de épocas de treinamento.
        """
        self.num_layers = len(layer_neurons)  #número total de camadas
        self.learning_rate = learning_rate #taxa de aprendizagem
        self.layer_neurons = layer_neurons  #array com os número de neurônio
        self.num_epochs = num_epochs #número de épocas
        self.smallest_error_computed_validation = float('inf') #menor erro computado para a validação durante o treinamento
        self.smallest_error_computed_validation_epoch = 0 #época do menor erro da validação
        self.smallest_error_computed_train = float('inf')
        self.smallest_error_computed_train_epoch = 0 #época do menor erro computado para o treino durante o treinamento
        self.accuracy_score = 0 #acurácia da rede (usar para crossValidation)
        self.mse_epoch_train = [] #armazena as épocas de treino
        self.mse_epoch_validation = [] #armazena as épocas de validação
        self.predicted_classes_initial_epoch=[] #classes apontadas no começo do treinamento
        self.predicted_classes_half_epoch=[] #classes preditas no meio do treinamento
        self.predicted_classes_final_epoch=[] #classes apontadas no fim do treinamento

        # Inicializa as camadas e seus atributos
        self.layers = {f"layer_{i+1}": {
            "neurons_num": layer_neurons[i],
            "output": None,
            "weight_matrix": None,
            "bias_matrix": None,
            "weight_correction": None,
            "bias_correction": None,
            "optimized_weight_matrix": None,
            "optimized_bias_matrix": None
        } for i in range(self.num_layers)}

    #inicializa os pesos e bias aleatoriamente entre -1 e 1
    def neural_network_init(self):
        """
        Inicializa os pesos e biases de todas as camadas, exceto a última.
        """
        for layer in range(1, self.num_layers):  # De layer_1 até a penúltima camada
            init_weight = weight_matrix_init(
                self.layers[f"layer_{layer}"]["neurons_num"],
                self.layers[f"layer_{layer + 1}"]["neurons_num"]
            )
            init_bias = bias_vector_init(self.layers[f"layer_{layer + 1}"]["neurons_num"])

            # Armazena os pesos e biases na camada correspondente
            self.layers[f"layer_{layer}"]["weight_matrix"] = init_weight
            self.layers[f"layer_{layer}"]["bias_matrix"] = init_bias


#classe que guarda info sobre a classe do problema a ser resolvido baseado no dataset de respostas
class NetworkClass:
    def __init__(self, neuralNetwork, y_dataset_raw, error_threshold):
      self.y_dataset_raw = y_dataset_raw.to_numpy()
      self.error_threshold = error_threshold
      self.num_possible_classes = len(np.unique(self.y_dataset_raw))

      if neuralNetwork.layers[f'layer_{neuralNetwork.num_layers}']['neurons_num'] == self.num_possible_classes:
        self.one_hot_matrix = np.eye(self.num_possible_classes)

      elif neuralNetwork.layers[f'layer_{neuralNetwork.num_layers}']['neurons_num'] == 1 and self.num_possible_classes == 2:
        self.one_hot_matrix = np.zeros((self.num_possible_classes, 1))
        self.one_hot_matrix[0][0] = 1

      else:
        last_layer_neurons_num = neuralNetwork.layers[f'layer_{neuralNetwork.num_layers}']['neurons_num']
        raise ValueError(
            f'ERRO - arquitetura da rede incompatível com o problema. há {self.num_possible_classes} classes, mas {last_layer_neurons_num} para representá-las.'
        )

      #resgata os valores únicos no y_dataset_raw de respostas para servir de rótulo
      self.classes = sorted(np.unique(self.y_dataset_raw))

      #cria um dicionário com os rótulos da classe como chave | classe 'A': [1, 0, 0]
      self.labeled_class = {label: self.one_hot_matrix[i] for i, label in enumerate(self.classes)}

      #cria um dicionário com os vetores onehot da classe como chave | [1, 0, 0]: classe 'A'
      self.enum_class = {tuple(vetor): label for label, vetor in self.labeled_class.items()}


      # Mapeamento índice | 0: [1, 0, 0]
      self.index_to_onehot = {i: self.one_hot_matrix[i] for i in range(self.num_possible_classes)}

      # Mapeamento vetor one-hot → índice (ex: [1, 0, 0] → 0)
      self.onehot_to_index = {tuple(vec): i for i, vec in self.index_to_onehot.items()}



    # transforma 0 => [1, 0, 0]
    def encode_index(self, index):
      return self.index_to_onehot[index]

    # transforma  [1, 0, 0] => 0
    def decode_index(self, onehot_vector):
      return self.onehot_to_index[tuple(onehot_vector)]

    #transforma [1, 0, 0] => 'A'
    def decode_class(self, predicted):
      #possível classe de acordo com o índice do maior valor
      possible_class = np.array(self.encode_index(np.argmax(predicted))).reshape(-1, 1)
      #np.all checa se a diferença entre a possível classe e o predito está abaixo do limite de erro
      if np.all(np.abs(predicted - possible_class) <= self.error_threshold):
        return self.enum_class[tuple(self.encode_index(np.argmax(predicted)))]
      else:
        return f'classe indefinida ou imprecisa; possível classe: {self.decode_class(possible_class)}'

    #transforma 'A' => [1, 0, 0]
    def encode_class(self, label):
      return self.labeled_class[label]



#função que separa um dataset em conjunto de treino, de teste e de validação
def train_test_validation_split(dataset, test_size):
  if test_size > 0.45:
    raise ValueError('Partição de teste muito grande para ser considerada. A rede neural precisa de espaço para treino')

  if test_size > 0.2:
    warnings.warn('Recomenda-se usar test_size até 0.15 para não prejudicar o conjunto de treino, pois tamanho da validação = tamanho de teste', UserWarning)

  dataset.rename(columns={dataset.columns[-1]: 'label'}, inplace=True)
  indices = dataset.index.tolist()
  test_indices = random.sample(population=indices, k=int(test_size * len(indices)))
  validation_indices = random.sample(population=(list(set(indices) - set(test_indices))), k=int(test_size * len(indices)))
  train_indices = list(set(indices) - set(test_indices) - set(validation_indices))

  test_df = dataset.loc[test_indices]
  validation_df = dataset.loc[validation_indices]
  train_df = dataset.loc[train_indices]

  return train_df, test_df, validation_df

def generate_confusion_matrix(networkClass, neuralNetwork, x_test, y_test):
  x_test = np.array(x_test)
  y_test = np.array(y_test)

  confusion_matrix = np.zeros((networkClass.num_possible_classes, networkClass.num_possible_classes))
  for row in range(0, x_test.shape[0]):
    actual_class = np.argmax(y_test[row]) #networkClass.decode_index(tuple(y_test[[row]].flatten()))

    predicted_value = predict(neuralNetwork, x_test[[row]].reshape(-1, 1), use_optimized_weights=True)

        #np.all(np.abs(predicted_value - y_test[[row]]) <= self.error_threshold) checa se os valores absolutos das diferenças são menores que o error_threshold
        #se for menor, logo o valor predito pertence a classe e adiciona +1 na célula correspondente na matriz de confusão
    if np.all(np.abs(predicted_value - y_test[[row]].reshape(-1, 1)) <= networkClass.error_threshold):
        confusion_matrix[actual_class][actual_class] += 1
    else:
        #predicted_class = networkClass.decode_index(tuple(np.round(predicted_value).flatten()))
        predicted_class = np.argmax(predicted_value)
        confusion_matrix[actual_class][predicted_class] += 1

  return confusion_matrix

#gera a matriz de pesos em [-1, 1]
def weight_matrix_init(rows, columns):
  matrix = np.random.uniform(-1, 1, (rows, columns))
  return matrix

#gera o vetor coluna de bias em [-1, 1]
def bias_vector_init(rows):
  matrix = np.random.uniform(-1, 1, (rows, 1))
  return matrix

#efetua a soma ponderada (w_ji*x_i + b) via multiplicacao de matrizes
def weighted_sum(weight_matrix, entry, bias):
  weight_dot_entry = np.dot(weight_matrix.T, entry)
  final_sum = weight_dot_entry + bias
  return final_sum

#passa a soma pela função de ativação e retorna a saída
def sigmoid_activation_function_output(sum):
  sum =  sum.astype(float)
  sum = np.clip(sum * -1, -700, 700)
  return 1/(1 + np.exp(sum))

#cálculo de custo, cálculo de erro
def error_function(target, final_output):
  return (np.mean(np.power((target - final_output), 2)))


#função de alimentação positiva da rede
def forward_pass(neuralNetwork, use_optimized_weights=False):
  if use_optimized_weights == True:
    weight_parameter = 'optimized_weight_matrix'
    bias_parameter = 'optimized_bias_matrix'
  else:
    weight_parameter = 'weight_matrix'
    bias_parameter = 'bias_matrix'

  layer_count = 2
  num_layers = neuralNetwork.num_layers
  while layer_count <= num_layers:
    sum_weights = weighted_sum(
        neuralNetwork.layers[f"layer_{layer_count - 1}"][weight_parameter],
        neuralNetwork.layers[f"layer_{layer_count - 1}"]["output"],
        neuralNetwork.layers[f"layer_{layer_count - 1}"][bias_parameter]
        )
    new_output = sigmoid_activation_function_output(sum_weights)
    neuralNetwork.layers[f"layer_{layer_count}"]["output"] = new_output

    layer_count += 1



#derivada da função de ativação
def sigmoid_activation_function_derivative(sum):
    return (sigmoid_activation_function_output(sum) * (1 - sigmoid_activation_function_output(sum)))



#algortimo de retropropagação do erro através da rede neural por completo
#armazenamento de pesos atualizados para update posterior
def backpropagation(neuralNetwork, target):
  num_layers_net = neuralNetwork.num_layers

  #cálculo do delta_k da camada de saída
  pre_activ = weighted_sum(
        neuralNetwork.layers[f"layer_{num_layers_net - 1}"]["weight_matrix"],
        neuralNetwork.layers[f"layer_{num_layers_net - 1}"]["output"],
        neuralNetwork.layers[f"layer_{num_layers_net - 1}"]["bias_matrix"]
        )
  small_delta_k = ((target - neuralNetwork.layers[f"layer_{num_layers_net}"]["output"]) * sigmoid_activation_function_derivative(pre_activ))


  #correção e armazenamento dos novos pesos da camada de sáida em forma de matriz
  weight_correction = np.dot(small_delta_k, neuralNetwork.layers[f"layer_{num_layers_net - 1}"]["output"].T)*neuralNetwork.learning_rate

  neuralNetwork.layers[f"layer_{num_layers_net - 1}"]["weight_correction"] = weight_correction.T

  bias_correction = neuralNetwork.learning_rate * small_delta_k
  neuralNetwork.layers[f"layer_{num_layers_net - 1}"]["bias_correction"] = bias_correction

  num_layers_net -= 1
  while (num_layers_net) > 1:
    pre_activ = weighted_sum(
        neuralNetwork.layers[f"layer_{num_layers_net - 1}"]["weight_matrix"],
        neuralNetwork.layers[f"layer_{num_layers_net - 1}"]["output"],
        neuralNetwork.layers[f"layer_{num_layers_net - 1}"]["bias_matrix"]
        )

    #cálculo do delta_j para correção de camadas escondidas
    small_delta_j_in = np.dot(neuralNetwork.layers[f"layer_{num_layers_net}"]["weight_matrix"], small_delta_k)
    small_delta_j = (small_delta_j_in * sigmoid_activation_function_derivative(pre_activ))

    #correção e armazenamento dos novos pesos das camadas escondidas
    weight_correction_hidden_layer = np.dot(small_delta_j, neuralNetwork.layers[f"layer_{num_layers_net - 1}"]["output"].T) * neuralNetwork.learning_rate
    neuralNetwork.layers[f"layer_{num_layers_net - 1}"]["weight_correction"] = weight_correction_hidden_layer.T

    bias_correction = neuralNetwork.learning_rate * small_delta_j
    neuralNetwork.layers[f"layer_{num_layers_net - 1}"]["bias_correction"] = bias_correction

    small_delta_k = small_delta_j

    num_layers_net -= 1



#atualiza os pesos após rodar toda a retropropagação do erro pela rede
def weight_update(neuralNetwork):
  num_layers = neuralNetwork.num_layers
  for layer in range(1, num_layers):
    neuralNetwork.layers[f"layer_{layer}"]["weight_matrix"] = (neuralNetwork.layers[f"layer_{layer}"]["weight_matrix"]) + (neuralNetwork.layers[f"layer_{layer}"]["weight_correction"])
    neuralNetwork.layers[f"layer_{layer}"]["bias_matrix"] = neuralNetwork.layers[f"layer_{layer}"]["bias_matrix"] + neuralNetwork.layers[f"layer_{layer}"]["bias_correction"]



#prediz valores baseados em input com os pesos já atualizados
def predict(neuralNetwork, input, use_optimized_weights=False):
  neuralNetwork.layers["layer_1"]["output"] = input
  forward_pass(neuralNetwork, use_optimized_weights)
  final_layer = neuralNetwork.num_layers
  return neuralNetwork.layers[f"layer_{final_layer}"]["output"]



#mede o erro em cada época da validação para calcular early stopping
def validation_mse_measurer(neuralNetwork, validation_input_data, validation_answer_data):
  num_layers_total = neuralNetwork.num_layers
  #lista que armazena os erros a cada batch de validação
  mse_list_per_row_validation = []
  validation_row = 0

  #laço que passa por todas as linhas de validação e armazena os erros quadráticos médios de cada linha de validação
  while validation_row < validation_input_data.shape[0]:
    #pred_validation guarda a sáida que a rede teve para a linha
    #error computed guarda o erro quadratico médio computada para a linha
    pred_validation = predict(neuralNetwork, validation_input_data.iloc[validation_row].to_numpy().reshape(-1, 1))
    error_computed = error_function(validation_answer_data.iloc[validation_row].to_numpy().reshape(-1, 1), pred_validation)

    mse_list_per_row_validation.append(error_computed)

    validation_row += 1

  #média dos erros quadráticos médios para obter o erro médio por época na validação
  mean_error_validation = np.mean(mse_list_per_row_validation)

  #retorna o valor do erro para uso no gráfico
  return mean_error_validation

#analisa se a rede está acertando os resultados para cada linha de validação para a estratégia cross-validation
def cross_validation_performance_measurer(neuralNetwork, validation_input_data, validation_answer_data, error_threshold):
  num_layers_total = neuralNetwork.num_layers

  #variável que conta os  acertos da rede para um conjunto de validação
  correct_answer_counter = 0

  validation_row = 0
  #laço que passa por todas as linhas de validação e armazena quantaz vezes a rede acertou em cada linha de validação
  while validation_row < validation_input_data.shape[0]:

    predicted_value = predict(neuralNetwork, validation_input_data.iloc[validation_row].values.reshape(-1, 1))
    answer = validation_answer_data.iloc[validation_row].values.reshape(-1, 1)

    metric = np.all(np.abs(predicted_value - answer) <= error_threshold)
    if metric:
      correct_answer_counter += 1


    validation_row += 1

  return correct_answer_counter


#executa o treinamento completo da rede neural
def train_network(neuralNetwork, train_dataset, answer_dataset, validation_input_dataset, validation_answer_dataset, error_threshold, cross_validation=False, early_stopping=True):
  #garante que não haja discrepância entre as linhas dos datasets
  if train_dataset.shape[0] != answer_dataset.shape[0] or validation_input_dataset.shape[0] != validation_answer_dataset.shape[0]:
    return "ERRO - número de linhas do dataset de treino ou validação e dataset de respostas não é equivalente. Revise todos os datasets"

  #número de linhas para cada época
  total_num_inputs = train_dataset.shape[0]

  #itera de acordo com o número de épocas
  counter = 0
  while counter != neuralNetwork.num_epochs:

    mse_list_per_row_train = []

    #itera sobre cada linha de treino/resposta
    for num_input in range(0, total_num_inputs):
      #remodela a linha de treino para garantir que tenha dimensão Lx1 e atribui a entrada a camada de entrada
      transformed_input = train_dataset.iloc[num_input].values.reshape(-1, 1)
      neuralNetwork.layers['layer_1']["output"] = transformed_input

      if(counter==0):
        #adiciona a classe apontada no inicio do treinamento
        neuralNetwork.predicted_classes_initial_epoch.append(predict(neuralNetwork, transformed_input))

      if(counter==neuralNetwork.num_epochs/2):
        #adiciona a classe apontada no meio do treinamento
        neuralNetwork.predicted_classes_half_epoch.append(predict(neuralNetwork, transformed_input))

      #faz o feedforward até a última camada
      forward_pass(neuralNetwork)

      #remodela a linha de reposta para garantir que tenha dimensão Lx1
      transformed_answer = answer_dataset.iloc[num_input].values.reshape(-1, 1)

      #cálculo e armazenamento de erro no treino para uso gráfico
      final_layer = neuralNetwork.num_layers

      #cálculo do MSE do treino
      #if early_stopping == True:
      error_train = error_function(transformed_answer, neuralNetwork.layers[f"layer_{final_layer}"]["output"])
      mse_list_per_row_train.append(error_train)

      #condição que analisa se a rede está convergindo com uma margem de erro
      smallest_error_computed_train_margin = (neuralNetwork.smallest_error_computed_train)*(0.9)
      if error_train < smallest_error_computed_train_margin:
        neuralNetwork.smallest_error_computed_train = error_train
        neuralNetwork.smallest_error_computed_train_epoch = counter

      #retropropaga o erro pela rede de acordo com a resposta desejada
      backpropagation(neuralNetwork, transformed_answer)

      #atualiza os pesos
      weight_update(neuralNetwork)

    if early_stopping == True:
      #salva o erro médio do treino na época vigente

      #faz o cálculo de erro da época para a validação
      validation_error = validation_mse_measurer(neuralNetwork, validation_input_dataset, validation_answer_dataset)

      #salva o erro médio da validação na época vigente
      neuralNetwork.mse_epoch_validation.append([counter, validation_error])

      #condição que analisa se o erro da validação na época vigente é menor que o menor erro de alguma época passada para gravar os pesos otimizados
      if validation_error < neuralNetwork.smallest_error_computed_validation:
      #salva o novo menor erro
        neuralNetwork.smallest_error_computed_validation = validation_error
        neuralNetwork.smallest_error_computed_validation_epoch = counter
        #salva os novos pesos otimizados
        for layer in range(1, neuralNetwork.num_layers):
          neuralNetwork.layers[f"layer_{layer}"]["optimized_weight_matrix"] = neuralNetwork.layers[f"layer_{layer}"]["weight_matrix"]
          neuralNetwork.layers[f"layer_{layer}"]["optimized_bias_matrix"] = neuralNetwork.layers[f"layer_{layer}"]["bias_matrix"]

    neuralNetwork.mse_epoch_train.append([counter, np.mean(mse_list_per_row_train)])

    #incrementa o contador para a próxima época de treinamento
    counter += 1

    #adiciona a classe apontada no fim do treinamento
  for i in range(total_num_inputs):
    input_sample = train_dataset.iloc[i].values.reshape(-1, 1)
    predicted_class = predict(neuralNetwork, input_sample)

    neuralNetwork.predicted_classes_final_epoch.append(predicted_class)


  if cross_validation == True:
    accuracy_score = cross_validation_performance_measurer(neuralNetwork, validation_input_dataset, validation_answer_dataset, error_threshold)
    neuralNetwork.accuracy_score = accuracy_score / validation_answer_dataset.shape[0]

  return f"Rede convergiu na época {neuralNetwork.smallest_error_computed_train_epoch} com erro de treino {neuralNetwork.smallest_error_computed_train:.6f}"

def k_fold_cross_validation(neuralNetwork, train_input_dataset, train_answer_dataset, k_fold_num, error_threshold):
  fold_section = train_input_dataset.shape[0]//k_fold_num

  layer_neurons = neuralNetwork.layer_neurons
  num_epochs = neuralNetwork.num_epochs
  learning_rate = neuralNetwork.learning_rate

  networks = {}
  for i in range(1, k_fold_num + 1):
    networks[f"network_{i}"] = NeuralNetwork(layer_neurons, learning_rate, num_epochs)
    networks[f"network_{i}"].neural_network_init()

  for n in range(1, k_fold_num + 1):
    validation_input = train_input_dataset.iloc[(n-1)*fold_section:n*fold_section]
    train_input = train_input_dataset.drop(train_input_dataset.index[(n-1)*fold_section:n*fold_section])

    validation_answer = train_answer_dataset.iloc[(n-1)*fold_section:n*fold_section]
    train_answer = train_answer_dataset.drop(train_answer_dataset.index[(n-1)*fold_section:n*fold_section])

    train_network(networks[f"network_{n}"], train_input, train_answer, validation_input, validation_answer, error_threshold, cross_validation=True, early_stopping=False)

  accuracy_list = []
  for net_num in range(1, k_fold_num + 1):
    acc = networks[f"network_{net_num}"].accuracy_score
    print(f'acurácia do fold {net_num} ==> {acc}')
    accuracy_list.append(acc)



  return np.array([[np.mean(accuracy_list)], [np.std(accuracy_list)]])
```

#Geração e visualização da matriz de confusão com heatmap

Neste caso está convencionado como:
- linha: representa a classe verdadeira
- coluna: representa o valor predito pelo modelo

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import string

cm = generate_confusion_matrix(nc, nn, x_test_df, y_test_df)
dcm = pd.DataFrame(cm)

labels = list(string.ascii_uppercase)

# Criar mapa de calor com cor única (por exemplo, azul)
sns.heatmap(dcm, cmap='gist_yarg', xticklabels=labels, yticklabels=labels, linewidths=0.5)  # você pode trocar 'Blues' por outra paleta


plt.gca().xaxis.set_ticks_position('top')
plt.gca().xaxis.set_label_position('top')



plt.title("Mapa de calor da matriz de confusão com o conjunto de testes | (tom mais forte = maior valor)", pad=25)

plt.show()

accuracy = np.trace(cm) / np.sum(cm)
print("\n")
print(f"acurácia medida pela matriz de confusão: {accuracy}")
print(f"erro pela matriz de confusão: {1 - accuracy}")
```
